```python
import pandas as pd
import numpy as np
```

```python
car = pd.read_csv('quikr_car.csv')
```

```python
car.head()
```

```python
backup = car.copy()
```

```python
car['year'].unique()
```

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype(int)
```

```python
car['year']
```

```python
car['Price']
```

```python
car = car[car['Price']!='Ask For Price']
```

```python
car['Price'] = car['Price'].str.replace(',','')
car['Price'] = car['Price'].astype(int)
```

```python
car['Price']
```

```python
car['kms_driven'].unique()
```

```python
car = car[car['kms_driven']!='Petrol']
```

```python
car['kms_driven'] = car['kms_driven'].str.replace(',','').str.replace(' kms','')
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car['kms_driven']
```

```python
car['company'].unique()
```

```python
car.dropna(inplace=True)
```

```python
car['name'].unique()
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car.head()
```

```python
car.reset_index(drop=True)
```

```python
car.to_csv('cleaned.csv')
```

```python
X = car.drop(columns='Price')
y = car['Price']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
y_train_pred = pipe.predict(X_train)
```

```python
r2_score(y_train,y_train_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mse = mean_squared_error(y_train, y_train_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_train, y_train_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
    
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
import matplotlib.pyplot as plt

indices = np.arange(len(y_test))

plt.figure(figsize=(10, 6))
plt.scatter(indices, y_test, color='blue', label='Actual', alpha=0.6)
plt.scatter(indices, y_pred, color='red', label='Predicted', alpha=0.6)

m, b = np.polyfit(indices, y_pred, 1)
plt.plot(indices, m * indices + b, color='green', linewidth=2, label='Best Fit Line (Predicted)')

plt.xlabel("Sample Index")
plt.ylabel("Price")
plt.title("Actual vs Predicted Prices")
plt.legend()
plt.grid(True)

plt.show()
```
